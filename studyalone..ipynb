## 파이썬에서 XML 데이터 다루기

```python
#XML 문자열을 변수에 저장
x_str="""
<book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
</book>
"""
```

```python
#문자열을 XML로 변환: fromstring() 함수
import xml.etree.ElementTree as et
book=et.fromstring(x_str)
```

```python
print(type(book))
#위 함수는 문자열을 파이썬 객체가 아닌 ElemenTree 모듈 아래에 정의된 Element 클래스의 객체로 반환됨을 확인
```

```python
print(book.tag)     #.tag 속성은 “이 엘리먼트의 이름”
```

```python
book_childs=list(book)    #book 객체를 리스트로 변환하여 자식 엘리먼트 구하기
print(book_childs)
```

```python
#위 리스트의 각 항목을 변후에 할당하고 text 속성으로 엘리먼트에 있는 텍스트 출력
name, author, year=book_childs
print(name.text)
print(author.text)
print(year.text)
```

### 정리:
name  : <name> 엘리먼트 객체

name.tag  : 'name'   (엘리먼트 이름)

name.text : '이지연' (엘리먼트 안의 값)

```python
#자식 엘리먼트 확인하기: findtext() 메서드
name=book.findtext('name')
author=book.findtext('author')
year=book.findtext('year')
print(name)
print(author)
print(year)
```

```python
#여러 개의 엘리먼트: 부모 엘리먼트 만들기
x2_str="""
<books>      #부모 엘리먼트
    <book>
        <name>혼자 공부하는 데이터 분석</name>
        <author>박해선</author>
        <year>2022</year>
    </book>
    <book>
        <name>혼자 공부하는 머신러닝+딥러닝</name>
        <author>박해선</author>
        <year>2020</year>
    </book>
</books>
"""
```

```python
books=et.fromstring(x2_str)    #부모엘리먼트 확인
print(books.tag)
```

```python
#동일한 이름을 가진 여러 개의 자식 엘리먼트 확인하기: findall() 메서드와 for문
for book in books.findall('book'):
    name=book.findtext('name')
    author=book.findtext('author')
    year=book.findtext('year')
    print(name)
    print(author)
    print(year)
    print()
```

```python
from bs4 import BeautifulSoup
```

```python
#이 클래스의 객체 생성하기. 첫 번째 매개변수는 파싱할 HTML 문서, 두 번쨰는 파싱에 사용할 파서
soup=BeautifulSoup(r.text, 'html.parser')
#파서: 입력 데이터를 받아 데이터 구조를 만드는 소프트웨어 라이브러리 (json 패키지, xml 패키지가 각각 JSON과 XML을 위한 파서라고 볼 수 있다.)
```

```python
#개발자 도구 창에서 찾은 링크는 HTML의 <a> 태그(마우스로 클릭할 수 있는 하이퍼링크를 만드는 태그) 안에 포함되어 있음.
```

```python
#태그 위치 찾기: find() 메서드 (soup 객체에게 </a> 태그의 위치를 찾으라고 지시하기)
#첫 번째 매개변수에는 찾을 태그 이름 지정, atters 매개변수에는 찾으려는 태그의 속성을 딕셔너리로 지정
#(예: soup.find('div', atters={'id':'search'})는 id 속성이 'search'인 <div> 태그를 찾으라는 의미)
prd_link=soup.find('a',attrs={'class':'gd_name'})     #<a> 태그 추출하기
```

```python
print(prd_link)    #링크가 포함되어 있는 prd_link는 뷰티플수프의 Tag 클래스 객체이다.
```

```python
#prd_link를 딕셔너리처럼 사용해 태그 안의 속성을 참조해서 링크 주소인 href 속성값 얻기
print(prd_link['href'])
```

```python
#성공입니다! 뷰티플수프를 사용해 Yes24의 검색 결과 페이지에서 도서의 상세 페이지로 넘어가는 링크를 찾았습니다. 
#이 주소만 있다면 다시 requests.get() 함수를 사용하여 쪽수가 담긴 상세 페이지 HTML을 가져올 수 있습니다.
```

```python
#도서 상세 페이지 HTML 가져오기
url='http://www.yes24.com' +prd_link['href']
r=requests.get(url)
```

```python
print(r.text)
```

```python
#도서 쪽수가 담긴 HTML 위치 찾기(크롬 브라우저 개발자 도구 열기)
soup=BeautifulSoup(r.text, 'html.parser')
prd_detail=soup.find('div',attrs={'id':'infoset_specific'})
print(prd_detail)
```

```python
#테이블 태그를 리스트로 가져오기:find_all() 메서드(특정 HTML 태그를 모두 찾아서 리스트로 반환)
#(품목정보 테이블의 행을 하나씩 검사해서 '쪽수, 무게, 크기'에 해당하는 <tr> 태그 찾기)
prd_tr_list=prd_detail.find_all('tr')
print(prd_tr_list)
#find() 메서드는 지정된 이름을 가진 첫 번쨰 태그를, find_all() 메서드는 지정된 이름을 가지니 모든 태그를 찾습니다.
```

```python
#태그 안의 텍스트 가져오기:get_text() 메서드
#for 문으로 prd_tr_list를 순회하면서 <th> 태그 안의 텍스트 검사, 원하는 행을 찾으면 변수에 저장
for tr in prd_tr_list:
    if tr.find('th').get_text()=='쪽수, 무게, 크기':
        page_td=tr.find('td').get_text()
        break
```

```python
#전체 도서의 쪽수 구하기(앞에서 했던 작업을 하나의 함수로 만들어 전체 판다스 데이터프레임에 적용)
def get_page_cnt(isbn):      #yes24 도서 검색 페이지 url
    url='http://www.yes24.com/Product/Search?doamin=BOOK&query={}'     #url에 isbn을 넣어 html 가져오기
    r=requests.get(url.format(isbn))    
    soup=BeautifulSoup(r.text, 'html.parser')    #html 파싱
    #검색 결과에서 해당 도서를 선책하기
    prd_info=soup.find('a',attrs={'class':'gd_name'})
    if prd_info ==None:
        return ''
    #도서 상세 페이지 가져오기
    url='http://www.yes24.com'+prd_info['href']
    r=requests.get(url)
    soup=BeautifulSoup(r.text, 'html.parser')     
    #품목정보 <div> 선택하기
    prd_detail=soup.find('div',attrs={'id':'infoset_specific'})
    #테이블에 있는 <tr> 태그 가져오기
    prd_tr_list=prd_detail.find_all('tr')
    #쪽수가 들어 있는 <th>를 찾아 <td>에 담긴 값을 반환하기
    for tr in prd_tr_list:
        if tr.find('th').get_text() == '쪽수, 무게, 크기':
            return tr.find('td').get_text().split()[0]
    return ''
```

### 일괄 처리 함수 만들기
-불필요한 데이터를 삭제하기 위한 코드들을 일괄 처리하는 data_cleaning()이라는 함수로 만들어 봅시다

```python
def data_cleaning(filename):
    """
    남산 도서관 장서 CSV 데이터 전처리 함수
    :param filename: CSV 파일 이름
    """
    ns_df=pd.read_csv(filename, low_memory=False)    #파일을 데이터프레임으로 읽습니다.
    ns_book=ns_df.dropna(axis=1, how='all')     #NaN인 열을 삭제합니다.
    count_df=ns_book[['도서명','저자','ISBN','권','대출건수']]    #대출건수를 합치기 위해 필요한 행만 추출하여 cound_df 데이터프레임을 만듭니다. 
    loan_count= count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()    #도서명, 저자, isbn, 권을 기준으로 대출건수를 groupby합니다.
    #원본 데이터프레임에서 중복된 행을 제외하고 고유한 행만 추출하여 복사합니다.
    dup_rows=ns_book.duplicated(subset=['도서명','저자','ISBN', '권'])
    unique_rows=~dup_rows
    ns_book3=ns_book[unique_rows].copy()
    #도서명, 저자, isbn, 권을 인덱스로 설정합니다.
    ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True)
    #loan_count에 저장된 누적 대출건수를 업데이트합니다.
    ns_book3.update(loan_count)
    #인덱스를 재설정합니다.
    ns_book4=ns_book3.reset_index()
    #원본 데이터프레임의 열 순서로 변경합니다.
    ns_book4=ns_book4[ns_book.columns]
    return ns_book4
```

```python
#위 함수가 제대로 작동하는지 확인- 원본 데이터인 csv 파일을 전달해 새로운 데이터프레임을 만든 후 ns_book4와 동일한지 비교
#다른 데이터프레임을 비교할 떄는 equals() 메서드를 사용합니다.
new_ns_book4=data_cleaning('ns_202512.csv')
ns_book4.equals(new_ns_book4)
```

### 정규표현식: 문자열 패턴을 찾아서 대체하기 위한 규칙의 모음

### 숫자 찾기: |d
-정규 표현식에서 숫자를 나타내는 기호는 |d입니다. 네 자리 연도에 해당하는 표현은 |d|d|d|d입니다. 표현식을 그룹으로 묶을 때는 괄호를 사용합니다. 뒤에 두 자리만 하나의 그룹으로 묶을 때는 |d|d(|d|d)처럼 씁니다.

```python
#'발행년도' 열의 값을 정규 표현식으로 두 자리 연도로 바꾸기
ns_book4.replace({'발행년도': {r'\d\d(\d\d)':r'\1'}}, regex=True)[100:102]     #패턴 안에 있는 첫 번째 그룹으로 연도를 바꿉니다.
#이때 정규 표현식 앞에 붙인 r문자는 파이썬에서 정규 표현식을 다른 문자열과 구분하기 위해 접두사처럼 붙입니다.
```

```python
#정규 표현식이 반복될 때는 일일이 쓰는 대신 중괄호를 사용하여 개수를 지정할 수 있습니다.(예: \d{2}와 \d\d는 같은 의미)
ns_book4.replace({'발행년도':{r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]   #위와 같은 코드
```

### 문자 찾기: 마침표(.)
어떤 문자에도 대응하는 정규 표현식 문자는 마침표(.)입니다. 또한 반복 개수를 지정하기 어려운 경우에는 *문자를 사용하여 0개 이상 반복된다고 표시할 수 있습니다.
정규 표현식에서 괄호는 그룹을 나타내는 데 사용하므로 일반 문자의 괄호라고 인식하게 하려면 역슬래시|를 앞에 붙어야 합니다. 그리고 띄어쓰기가 있으므로 공백 문자를 나타내는 정규 표현식 |s를 앞에 붙입니다.

```python
ns_book4.replace({'저자':{r'(.*)\s\(지은이\)(.*)\s\(옮긴이\)':r'\1\2'}, '발행년도': {r'\d{2}(\d{2})':r'\1'}}, regex=True)[100:102]
#1. 저자 이름(.*)- 살리는데 몇글잔지 몰라서 *   2. (지은이) 찾아 삭제 \s\(지은이\)   3. 쉼표와 역자 이름은 살리기   4. 옮긴이 삭제
```

```python
#도서명을 가져오는 함수
def get_book_title(isbn):    #yes24 도서 검색 페이지 url
    url='http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    #url에 isbn을 넣어 html 가져옵니다.
    r=requests.get(url.format(isbn))
    soup=BeautifulSoup(r.text, 'html.parser')    #html 파싱

    print(soup.select_one('.gd_name'))

    #클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져옵니다.
    title = soup.select_one('.gd_name').get_text()
    #title= soup.find('a', attrs={'class': 'gd_name'}).get_text()를 위 코드로 고쳤습니다.
    return title
```

```python
#도서명을 추출한 것과 같은 방식으로 저자, 출판사, 발행 연도를 추출하여 반환하는 함수를 만들어 보겠습니다.
#도서명과 달리 저자는 두 명 이상일 수 있기 때문에 뷰티플수프의 find_all() 메서드를 사용해 저자를 담은 <a>태그를 모두 추출해 파이썬 리스트에 저장
#그 다음 추출한 결과를 join() 메서드를 사용해 하나의 문자열로 합쳐 줍니다.
```

```python
#발행 연도는 정규식을 사용하여 월을 제외해 연도만 추출해야 합니다. 파이썬에거 정규 표현식을 지원하는 re모듈의 findall() 함수를 사용하면 원하는 정규식에 매칭되는 모든 문자열을 찾아 리스트로 반환해 줍니다.
```

```python
import re

def get_book_info(row):
    title=row['도서명']
    author=row['저자']
    pub=row['출판사']
    year=row['발행년도']
    #Yes24 도서 검색 페이지 URL
    url='http://www.yes24.com/Product/Search?domain=BOOK&query={}'
    #URL에 ISBN을 넣어 html 가져옵니다.
    r=requests.get(url.format(row['ISBN']))
    soup=BeautifulSoup(r.text, 'html.parser')    #html 파싱
    try:
        if pd.isna(title):
            #클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져옵니다.
            title=soup.find('a', attrs={'class':'gd_name'}).get_text()
    except AttributeError:
        pass

    try:
        if pd.isna(author):
            #클래스 이름이 'info_auth'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옵니다.
            authors=soup.find('span', attrs={'class':'info_auth'}).find_all('a')
            author_list=[auth.get_text() for auth in authors]
            author=','.join(author_list)
    except AttributeError:
        pass
    try:
        if pd.isna(pub):
            #클래스 이름이 'info_pub'인 <span> 태그 아래 <a> 태그의 텍스트를 가져 옵니다.
            pub=soup.find('span', attrs={'class':'info_pub'}).find('a').get_text()
    except AttributeError:
        pass

    try:
        if year == -1:
            #클래스 이름이 'info_date'인 <span> 태그 아래 텍스트를 가져옵니다.
            year_str=soup.find('span', attrs={'class':'info_date'}).get_text()
            #정규 표현식으로 찾은 값 중에 첫 번째 것만 사용합니다.
            year=re.findall(r'\d{4}', year_str)[0]
    except AttributeError:
        pass

    return title, author, pub, year

#이 함수는 누락된 값에만 뷰티플수프로 추출한 값을 저장합니다. 만약 뷰티플수프로 추출할 수 없는 경우(yes24에 도서 정보가 없거나, html 요소가 누락된 경우)에는 오류 발생
```

```python
#누락된 값이 있었던 처음 두 개의 행에 방금 작성한 get_book_info() 함수를 적용해 보겠습니다. 
#함수가 여러 개의 값을 반환하는 경우 apply() 메서드는 기본적으로 반환된 값을 하나의 튜플로 만듭니다. 
#따라서 reqult_type 매개변수를 'expand'로 지정하여 반환된 값을 각기 다른 열로 만들겠습니다.
updated_sample=ns_book5[na_rows].head(2).apply(get_book_info, axis=1, result_type='expand')
updated_sample
```

```python
#그리고 ns_book5 데이터프레임을 ns_book5_update 데이터프레임 데이터로 업데이터한 후 누락된 행이 몇 개인지 다시 확인해 보겠습니다.
# CSV 업데이트 df로 결측만 채우기 (ns_book5_update 사용!)
for col in ns_book5_update.columns:
    ns_book5[col] = ns_book5[col].fillna(ns_book5_update[col])

#누락 체크
na_rows = (
    ns_book5["도서명"].isna()
    | ns_book5["저자"].isna()
    | ns_book5["출판사"].isna()
    | ns_book5["발행년도"].eq(-1)
)
print(na_rows.sum())
```

```python
#누락된 값이 있는 행이 6447에서 6257로 줄었습니다! 하하 이제 마지막으로 누락된 값을 가진 행을 삭제하여 분석 대상에서 제외하겠습니다. 
#(방법은 맞는데 실제로 채울 데이터가 이번 케이스에는 없는가보네용)
#dropna() 메서드에 '도서명', '저자','출판사' 열을 리스트로 지정한 후 누락된 값이 있는 행을 삭제하고 그다음 '발행년도' 열 값이 -1이 아닌 행만 선택해 프레임 생성
# 1️⃣ 발행년도 정리 (−1 → NA 통일)
ns_book5["발행년도"] = ns_book5["발행년도"].replace(-1, pd.NA)
# 2️⃣ 핵심 컬럼 결측 제거
ns_book6 = ns_book5.dropna(subset=["도서명", "저자", "발행년도"])  #추가 팁- 혹시 출판사 NA가 너무 많아서 데이터가 많이 날아가고 출판사 없이 분석해도 되는 경우라면 이렇게 제거.
ns_book6 = ns_book6.dropna(subset=["출판사"])     # 출판사도 결측 제거
ns_book6["발행년도"] = ns_book6["발행년도"].astype("int32")  # 발행년도 정수형 변환- 이 줄은 맨 마지막!
ns_book6.head()
```
